## Setting Up Your Python Environment

Before you begin, you'll need to set up your Python environment. This section will guide you through the installation process.

### Installing Python

If you don't already have Python installed, you can download it from the official website: [python.org](https://www.python.org/downloads/)

We recommend installing Python 3.8 or later. During installation:
- On Windows, make sure to check the box that says "Add Python to PATH"
- On macOS and Linux, Python might already be installed, but you may want to install a newer version

To verify your installation, open a terminal or command prompt and type:
```
python --version
```
or
```
python3 --version
```

### Installing Jupyter Notebook

These workshop materials are provided as Jupyter Notebooks. Jupyter Notebook is an open-source web application that allows you to create and share documents containing live code, equations, visualizations, and narrative text.

There are several ways to install Jupyter Notebook:

#### Option 1: Using pip (Python's package installer)

Open a terminal or command prompt and run:
```
pip install notebook
```
or
```
pip3 install notebook
```

#### Option 2: Using Anaconda (Recommended for beginners)

Anaconda is a distribution of Python that comes with many packages pre-installed, including Jupyter Notebook. It's particularly useful for data science and machine learning.

1. Download Anaconda from [anaconda.com](https://www.anaconda.com/products/individual)
2. Follow the installation instructions for your operating system
3. Once installed, you can launch Jupyter Notebook from the Anaconda Navigator or by typing `jupyter notebook` in your terminal or command prompt

### Installing Required Packages

For this workshop, you'll need several Python packages, especially for the data visualization section. You can install them using pip or conda (if you're using Anaconda).

#### Using pip:

```
pip install numpy pandas matplotlib seaborn
```

#### Using conda (if you installed Anaconda):

```
conda install numpy pandas matplotlib seaborn
```

These packages provide the following functionality:
- **NumPy**: Support for large, multi-dimensional arrays and matrices, along with mathematical functions to operate on these arrays
- **Pandas**: Data manipulation and analysis library that provides data structures like DataFrames
- **Matplotlib**: Comprehensive library for creating static, animated, and interactive visualizations
- **Seaborn**: Statistical data visualization library based on Matplotlib that provides a high-level interface for drawing attractive graphics

### Verifying Your Installation

Let's verify that everything is installed correctly by importing the required packages:

````python
# Import the required packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Print the versions
print(f"NumPy version: {np.__version__}")
print(f"Pandas version: {pd.__version__}")
print(f"Seaborn version: {sns.__version__}")

# Create a simple plot to verify that visualization works
plt.figure(figsize=(6, 4))
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.title("Sine Wave")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.grid(True)
plt.show()

print("\nIf you can see a sine wave above, your installation is working correctly!")
````

## Jupyter Notebook Basics

If you're new to Jupyter Notebook, here are some basic commands to help you get started:

- **Create a new cell**: Click the "+" button in the toolbar
- **Run a cell**: Select the cell and press Shift+Enter
- **Change cell type**: Use the dropdown menu in the toolbar to change between Code and Markdown
- **Save the notebook**: Click the save button or press Ctrl+S (Cmd+S on Mac)
- **Add a new notebook**: Click "New" in the Jupyter dashboard and select "Python 3"
- **Keyboard shortcuts**: Press "H" in command mode to see a list of keyboard shortcuts

Jupyter Notebook has two modes:
- **Command mode**: Press Esc to enter command mode. In this mode, you can navigate between cells using the arrow keys.
- **Edit mode**: Press Enter to enter edit mode. In this mode, you can edit the content of a cell.
